```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
#plt.style.use('ggplot')
%matplotlib inline
```

```python
from BJH_function.BJH_calculation import BJH_method as BJH
```

```python
"""Reading pressure and adsorption quantity data"""
#filename = 'pyrolysis_2_93.xlsx'
filename ='test_samples/test_iso.xlsx'
sheetname = '3_14_F1_N2'
gas_type = 'N2'
use_pressure = True
iso_df = pd.read_excel(filename,sheetname=sheetname)
p_measure = np.array(iso_df['p_rels'])
q_measure = np.array(iso_df['q_abs'])

# read psd
filename, sheetname = 'test_samples/test_psd.xlsx', '3_14_F1_N2'
psd_df = pd.read_excel(filename,sheetname)
D_3flex, PV_3flex = psd_df['D'].values, psd_df['IPV'].values 
```

```python
"""Use BJH calculate PSD"""
my_BJH = BJH(use_pressure=use_pressure,gas_type=gas_type)
my_BJH.fit(p_measure,q_measure)
my_BJH.plot_isotherm()
ax = my_BJH.plot_BJH_psd()
ax.plot(D_3flex, PV_3flex,'ob--', label = '3flex')
ax.legend()
```
